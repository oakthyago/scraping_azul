```python
import time

import re
import pyautogui as gui
import pyperclip
from datetime import datetime, timedelta
import sys

# ====== PARÂMETROS ======
LIMITE_PONTOS = 25000
PAUSA_CURTA = 1.0   
PAUSA_MEDIA = 1.5
PAUSA_LONGA = 5

# Garante uma pequena pausa padrão entre ações do pyautogui
gui.PAUSE = 0.05

# ====== 1) Fluxo inicial ======
# 820 400
gui.click(820, 400)
time.sleep(5)  # TIME WAIT 3S

# 500 400 / BACKSPACE
gui.click(500, 400); time.sleep(2)
gui.press('backspace')

# 813 400 / BACKSPACE
gui.click(813, 400); time.sleep(2)
gui.press('backspace')

# 500 400 / TYPE RECIFE
gui.click(500, 400); time.sleep(2)
gui.typewrite('RECIFE', interval=0.05)
time.sleep(2)  # TIM25EWAIT(3S)
gui.press('enter')

# type GIG
gui.click(770, 400); time.sleep(PAUSA_CURTA)    
gui.typewrite('GIG', interval=0.05)
time.sleep(2)  # TIMEWAIT(3S)
gui.press('enter')

# ====== 2) Loop de datas: 25/09/2025 até 13/10/2025 (ddmmaaaa) ======
inicio = datetime(2025, 9, 25)
fim    = datetime(2025, 10, 13)

# Regex: pega números com milhar (ex.: 10.000) seguidos de 'pontos'
re_pontos = re.compile(r'\b(\d{1,3}(?:\.\d{3})+|\d+)\b\s*pontos', re.IGNORECASE)

resultados = []  # para log

dia = inicio
while dia <= fim:
    data_str = dia.strftime('%d%m%Y')  # ddmmaaaa

    # 820 400 / backspace 8x / digita data
    gui.click(1020, 400); time.sleep(PAUSA_CURTA)
    for _ in range(8):
        gui.press('backspace'); time.sleep(0.02)
    gui.typewrite(data_str, interval=0.04)

    # 1200 900 (talvez 'buscar')
    gui.click(1281, 842); time.sleep(6)

    # 1000 830 (talvez confirmar/abrir resultados)
    gui.click(1000, 830)
    time.sleep(10)  # TIME WAIT 10S (carregar resultados)

    # PRESS END / 950 826 / TIME WAIT 5S / PRESS END / 950 826
    gui.press('end'); time.sleep(PAUSA_MEDIA)
    gui.click(950, 826); time.sleep(5)  # TIME WAIT 5S
    gui.press('end'); time.sleep(PAUSA_CURTA)
    gui.click(950, 826); time.sleep(PAUSA_CURTA)

    # 1760 630 / CTRL+A / CTRL+C
    gui.click(1760, 630); time.sleep(PAUSA_CURTA)
    gui.hotkey('ctrl', 'a'); time.sleep(PAUSA_CURTA)
    gui.hotkey('ctrl', 'c'); time.sleep(PAUSA_CURTA)
    gui.hotkey('home'); time.sleep(2)
    gui.click(820, 400)
    time.sleep(5)  # TIME WAIT 3S

    texto = pyperclip.paste() or ""

        # Normaliza quebras pra linha única
    texto_lin = texto.replace('\r\n', '\n')

    # Casa SOMENTE linhas que são "NN.NNN pontos" (ou "25000 pontos"), sozinhas na linha
    import re
    re_linha_pontos = re.compile(r'(?mi)^\s*(\d{1,3}(?:\.\d{3})+|\d+)\s*pontos\s*$')
    
    encontrados = re_linha_pontos.findall(texto_lin)  # sem drop do primeiro

    encontrados_validos = []
    for num in encontrados:
        val = int(num.replace('.', ''))
        if val <= LIMITE_PONTOS:
            encontrados_validos.append(val)
    # Log / feedback
    if encontrados_validos:
        # bip
        try:
            import winsound
            winsound.Beep(1000, 300)
        except Exception:
            sys.stdout.write('\a'); sys.stdout.flush()

        print(f"{data_str}: ENCONTRADO <= {LIMITE_PONTOS} -> {sorted(encontrados_validos)}")
        resultados.append((data_str, sorted(encontrados_validos)))
    else:
        print(f"{data_str}: nada <= {LIMITE_PONTOS}")

    # Próximo dia
    
    if dia == fim:
        dia = inicio
    else:
        dia += timedelta(days=1)
    time.sleep(0.3)

print("\n=== RESUMO ===")
for d, vals in resultados:
    print(f"{d}: {vals}")
```
